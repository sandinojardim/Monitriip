```python
import pandas as pd
vg = pd.read_csv('viagem_regular_2022.csv',encoding = "latin",sep=',')
#bg_vg = vg[(vg['cnpj'] == 3143492000162)]
```

```python
mask = (vg['latitude'] >= -15.9658) & (vg['latitude'] <= -15.8435) & (vg['longitude'] >= -52.3678) & (vg['longitude'] <= -52.2044)
bg_vg = vg[mask]
```

```python
bg_vg['cnpj'].value_counts()
```

```python
import folium
from folium.plugins import MiniMap

# Create a DataFrame with latitudes and longitudes
df = pd.DataFrame({'lat': bg_vg['latitude'], 'long': bg_vg['longitude']})

# Create a map centered on the first coordinate
map = folium.Map(location=['-15.88', '-52.27'], zoom_start=10)

countIda = 0
countVolta = 0
# Add markers for each coordinate
for index, row in df.iterrows():
    
    if bg_vg.at[index, 'sentido_linha'] == 0:
        color = 'red'
        countIda += 1
    else:
        color = 'blue'
        countVolta +=1
    
    folium.Marker([row['lat'], row['long']],icon=folium.Icon(color=color)).add_to(map)
    
# Add a control showing the number of markers
folium.map.Marker(
    ['-15.88', '-52.27'],
    icon=folium.Icon(color='white', icon_color='black', icon='info-sign'),
    popup=f'Number of markers: {len(df)}\n Ida:{countIda}\nVolta:{countVolta}',
).add_to(map)

# Add a minimap control
minimap = MiniMap(tile_layer='Stamen Terrain')
map.add_child(minimap)
# Display the map
map
```
